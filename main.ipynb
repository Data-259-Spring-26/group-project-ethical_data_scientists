```python
import requests
from urllib.parse import urlparse

GAMMA = "https://gamma-api.polymarket.com"

url = "https://polymarket.com/event/cbb-vcu-ncar-2026-03-19"
slug = urlparse(url).path.strip("/").split("/")[-1]

event = requests.get(f"{GAMMA}/events/slug/{slug}").json()

for market in event["markets"]:
    print(market["id"], "-", market["question"])
```

```python
import requests
import json

GAMMA = "https://gamma-api.polymarket.com"

# Get market details
market_id = "1602050"
response = requests.get(f"{GAMMA}/markets/{market_id}")

if response.status_code == 200:
    market = response.json()
    print("Market:", market['question'])
    print("Closed:", market['closed'])
    print("Final Outcome Prices:", market['outcomePrices'])
    print("Outcomes:", market['outcomes'])
    print("Last Trade Price:", market.get('lastTradePrice'))
    print("\nToken IDs:", market.get('clobTokenIds'))
    print("\nFull market data:")
    print(json.dumps(market, indent=2))
else:
    print(f"Error: {response.status_code}")

# Note: The CLOB prices-history endpoint doesn't retain historical price data
# after markets settle. For historical tick data, you would need to:
# 1. Query while the market is active
# 2. Use a third-party Polymarket data archive
# 3. Use Gamma API to get snapshots if available
```
